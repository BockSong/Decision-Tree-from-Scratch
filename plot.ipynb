```python
from matplotlib import pyplot as plt
from decisionTree import *

train_file = "politicians_train.tsv"
test_file = "politicians_test.tsv"
train_out = "_2_.labels"
test_out = "2.labels"

def draw_plot(iter_num):
    train_error, test_error = [], []
    for iter_depth in range(iter_num):
        dt = decision_tree(iter_depth)
        dt.build_tree(train_file)
        train_error.append(dt.evaluate(train_file, train_out))
        test_error.append(dt.evaluate(test_file, test_out))

    depth_num = list(range(iter_num))
    #depth_num = [i + 1 for i in depth_num]
    
    plt.title("Error-Depth Plot on Politicians Dataset") 
    plt.xlabel("Depth of the tree") 
    plt.ylabel("Error") 
    plt.plot(depth_num, train_error, label = "train error") 
    plt.plot(depth_num, test_error, label = "test error") 
    plt.legend()
    plt.show()
```

```python
model = decision_tree(2)
model.build_tree(train_file)

draw_plot(len(model.dataset[0]))
```
